```python
total=data1.isnull().sum().sort_values(ascending=False)
per=(data1.isnull().sum()/data1.isnull().count()*100).sort_values(ascending=False)
pd.concat([total,per],keys=['Missing','Percentage'],axis=1)
```

```python
col=per[per>20].keys()
```

```python
col
```

```python
data2=data1.drop(col,axis=1)
```

```python
data2.shape
```

```python
data3=data2.select_dtypes(include=['int64','float64'])
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data3.isnull())
```

```python
missing_var=[var for var in data3.columns if data3[var].isnull().sum()>0]
```

```python
plt.figure(figsize=(16,9))
sns.set()
for i ,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data3[var])
```

```python
df4_num=data3.fillna(data3.mean())
```

```python
df4_num.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
sns.set()
for i,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data3[var],label="Original",color='red')
    sns.distplot(df4_num[var],label="after applying",color="blue")
```
