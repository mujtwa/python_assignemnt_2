```python
def remove_above_79_permissing(df):
    list=[]
    for var in df.columns:
        if df[var].isnull().sum()>690:
            list.append(var)
    return list        
            
            
```

```python
def fill_missing_all(df):
    for var in df.columns:
        if df[var].dtypes == ('int64' or 'float64'):
            df[var]=df[var].fillna(df[var].median())
        else:
            df[var]=df[var].fillna(df[var].dropna().mode().values[0])
            
```

```python
fill_missing_all(train)
```
