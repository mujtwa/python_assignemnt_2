```python
data=pd.read_csv("https://raw.githubusercontent.com/robertmartin8/MachineLearningStocks/master/stock_prices.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
pd.set_option("max.columns",None)
pd.set_option("max.rows",None)
```

```python
data.isnull().sum()
```

```python
total=data.isnull().sum().sort_values(ascending=False)
per=(data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
```

```python
pd.concat([total,per],keys=['Total_Missing','Percentage'],axis=1)
```

```python
data.columns
```

```python

col_del=[column for column in data.columns if (data[column].isnull().sum())>1990]
```

```python
data.drop(col_del,axis=1,inplace=True)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
thres=VarianceThreshold(threshold=0)
```

```python
thres.fit(data)
```

```python
thres.get_support()
```

```python
 data.columns[thres.get_support()]
```

```python
constant_columns=[column for column in data.columns if column not in data.columns[thres.get_support()]]
```
