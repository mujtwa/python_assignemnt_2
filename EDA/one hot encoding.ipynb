```python
top_10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).keys()]
# .index
```

```python
# and now how we make the 10 binary variables
```

```python
for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)
data[['X2']+top_10].head(40)    
```

```python
# get whole set of dummy variables, for all the categorical varibles
def one_hot_top_x(df,varible,top_x_labels):
    #function to create the dummy varibles for the most frequent labels
    #we can vary the number of most frequent labels that we encode
    for label in top_x_labels:
        df[varible+'_'+label]=np.where(data[varible]==label,1,0)
```

```python
# encode x2 into the 10 most frequent categories
```

```python
one_hot_top_x(data,'X2',top_10)
```
