```python
def sample_means(population,sample_quentity,sample_size):
    samples=[]
    for i in range(sample_quentity):
        samples1= population.sample(n=sample_size)
        sample_mean=np.array(samples1).mean()
        samples.append(sample_mean)
    return samples    
```

```python
def sampling(p,s_qnt,s_size):
    sample_means=[]
    for i in range(s_qnt):
        sample_size=p.sample(n=s_size)
        sample_mean=sample_size.mean()
        sample_means.append(sample_mean)
    return pd.Series(sample_means)
        
```

```python
array=np.random.rand(1000)
```

```python
Series_rand=pd.Series(array)
```

```python
sampling(Series_rand,500,200).hist()
```
